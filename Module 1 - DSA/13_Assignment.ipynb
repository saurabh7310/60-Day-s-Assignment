```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def createNewList(list1, list2):
    if not list1 or not list2:
        return None

    head = None
    current = None

    while list1 and list2:
        if list1.val >= list2.val:
            new_node = ListNode(list1.val)
            list1 = list1.next
        else:
            new_node = ListNode(list2.val)
            list2 = list2.next

        if not head:
            head = new_node
            current = new_node
        else:
            current.next = new_node
            current = current.next

    if list1:
        current.next = list1
    if list2:
        current.next = list2

    return head
# Create the first linked list based on the input values
list1 = ListNode(5)
list1.next = ListNode(2)
list1.next.next = ListNode(3)
list1.next.next.next = ListNode(8)

# Create the second linked list based on the input values
list2 = ListNode(1)
list2.next = ListNode(7)
list2.next.next = ListNode(4)
list2.next.next.next = ListNode(5)

# Create the new linked list by selecting the greater nodes
new_list = createNewList(list1, list2)

# Print the new linked list
current = new_list
while current:
    print(current.val, end=" ")
    current = current.next
# Output: 5 7 4 8

print()

# Create the first linked list based on the input values
list1 = ListNode(2)
list1.next = ListNode(8)
list1.next.next = ListNode(9)
list1.next.next.next = ListNode(3)

# Create the second linked list based on the input values
list2 = ListNode(5)
list2.next = ListNode(3)
list2.next.next = ListNode(6)
list2.next.next.next = ListNode(4)

# Create the new linked list by selecting the greater nodes
new_list = createNewList(list1, list2)

# Print the new linked list
current = new_list
while current:
    print(current.val, end=" ")
    current = current.next
# Output: 5 8 9 4
```

## Question 2

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def removeDuplicates(head):
    if head is None:
        return head

    current = head

    while current.next:
        if current.val == current.next.val:
            current.next = current.next.next
        else:
            current = current.next

    return head
# Create the linked list based on the input values
head = ListNode(11)
head.next = ListNode(11)
head.next.next = ListNode(11)
head.next.next.next = ListNode(21)
head.next.next.next.next = ListNode(43)
head.next.next.next.next.next = ListNode(43)
head.next.next.next.next.next.next = ListNode(60)

# Remove duplicate nodes from the linked list
modified_head = removeDuplicates(head)

# Print the modified linked list
current = modified_head
while current:
    print(current.val, end=" ")
    current = current.next
# Output: 11 21 43 60
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def reverseKGroup(head, k):
    if head is None or k == 1:
        return head

    dummy = ListNode(0)
    dummy.next = head
    prev = dummy

    while True:
        group_head = prev.next
        tail = prev

        for _ in range(k):
            tail = tail.next
            if tail is None:
                return dummy.next

        next_group_head = tail.next
        tail.next = None
        prev.next = reverse(group_head)
        group_head.next = next_group_head
        prev = group_head

    return dummy.next

def reverse(head):
    prev = None
    curr = head

    while curr:
        next_node = curr.next
        curr.next = prev
        prev = curr
        curr = next_node

    return prev
# Create the linked list based on the input values
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(2)
head.next.next.next = ListNode(4)
head.next.next.next.next = ListNode(5)
head.next.next.next.next.next = ListNode(6)
head.next.next.next.next.next.next = ListNode(7)
head.next.next.next.next.next.next.next = ListNode(8)

# Reverse every k nodes in the linked list
k = 4
modified_head = reverseKGroup(head, k)

# Print the modified linked list
current = modified_head
while current:
    print(current.val, end=" ")
    current = current.next
# Output: 4 2 2 1 8 7 6 5
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def reverseAlternateKNodes(head, k):
    if head is None or k == 1:
        return head

    current = head
    prev = None
    count = 0

    # Reverse the first k nodes
    while current and count < k:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node
        count += 1

    # Recursive call to reverse the next k nodes
    if count >= k:
        head.next = reverseAlternateKNodes(current, k)

    # Connect the reversed sublist with the remaining list
    else:
        head.next = current

    # Skip the next k nodes to maintain the alternating pattern
    count = 0
    while count < k-1 and current:
        current = current.next
        count += 1

    # Recursive call to reverse the alternate k nodes
    if current:
        current.next = reverseAlternateKNodes(current.next, k)

    return prev
# Create the linked list based on the input values
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)
head.next.next.next.next = ListNode(5)
head.next.next.next.next.next = ListNode(6)
head.next.next.next.next.next.next = ListNode(7)
head.next.next.next.next.next.next.next = ListNode(8)
head.next.next.next.next.next.next.next.next = ListNode(9)

# Reverse every alternate k nodes in the linked list
k = 3
modified_head = reverseAlternateKNodes(head, k)

# Print the modified linked list
current = modified_head
while current:
    print(current.val, end=" ")
    current = current.next
# Output: 3 2 1 4 5 6 9 8 7
```

## Question 5

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def deleteLastOccurrence(head, key):
    if head is None:
        return head

    last_occurrence = None
    prev_last_occurrence = None
    current = head
    prev = None

    # Find the last occurrence of the key
    while current:
        if current.val == key:
            last_occurrence = current
            prev_last_occurrence = prev
        prev = current
        current = current.next

    # If no occurrence is found, return the original list
    if last_occurrence is None:
        return head

    # If the last occurrence is the head, update the head pointer
    if last_occurrence == head:
        head = head.next
    else:
        prev_last_occurrence.next = last_occurrence.next

    return head
# Create the linked list based on the input values
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(5)
head.next.next.next.next = ListNode(2)
head.next.next.next.next.next = ListNode(10)

# Delete the last occurrence of the key in the linked list
key = 2
modified_head = deleteLastOccurrence(head, key)

# Print the modified linked list
current = modified_head
while current:
    print(current.val, end=" ")
    current = current.next
# Output: 1 2 3 5 10
```

## Question 6

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def mergeTwoLists(a, b):
    if a is None:
        return b
    if b is None:
        return a

    # Determine the head of the merged list
    if a.val <= b.val:
        head = a
        a = a.next
    else:
        head = b
        b = b.next

    # Merge the remaining nodes
    current = head

    while a and b:
        if a.val <= b.val:
            current.next = a
            a = a.next
        else:
            current.next = b
            b = b.next
        current = current.next

    # Append the remaining nodes from either list
    if a:
        current.next = a
    if b:
        current.next = b

    return head
# Create the first linked list based on the input values
a = ListNode(5)
a.next = ListNode(10)
a.next.next = ListNode(15)

# Create the second linked list based on the input values
b = ListNode(2)
b.next = ListNode(3)
b.next.next = ListNode(20)

# Merge the two linked lists
merged_head = mergeTwoLists(a, b)

# Print the merged linked list
current = merged_head
while current:
    print(current.val, end=" ")
    current = current.next
# Output: 2 3 5 10 15 20
```

## Question 7

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

def reverseDoublyLinkedList(head):
    if head is None:
        return None

    current = head

    # Swap prev and next pointers of each node
    while current:
        temp = current.prev
        current.prev = current.next
        current.next = temp
        current = current.prev

    # Update the new head of the reversed list
    head = temp.prev

    return head
# Create the doubly linked list based on the input values
head = Node(10)
node1 = Node(8)
node2 = Node(4)
node3 = Node(2)

head.next = node1
node1.prev = head
node1.next = node2
node2.prev = node1
node2.next = node3
node3.prev = node2

# Reverse the doubly linked list
reversed_head = reverseDoublyLinkedList(head)

# Print the reversed linked list
current = reversed_head
while current:
    print(current.data, end=" ")
    current = current.next
# Output: 2 4 8 10
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

def deleteNode(head, position):
    if head is None:
        return None

    # Special case: Delete the head node
    if position == 1:
        head = head.next
        if head:
            head.prev = None
        return head

    current = head
    count = 1

    # Traverse to the node at the given position
    while current and count < position:
        current = current.next
        count += 1

    # If the node at the given position is not found, return the original list
    if current is None:
        return head

    # Update the previous node's next pointer
    current.prev.next = current.next

    # Update the next node's previous pointer
    if current.next:
        current.next.prev = current.prev

    return head
# Create the doubly linked list based on the input values
head = Node(1)
node1 = Node(3)
node2 = Node(4)

head.next = node1
node1.prev = head
node1.next = node2
node2.prev = node1

# Delete the node at position 3 in the doubly linked list
position = 3
modified_head = deleteNode(head, position)

# Print the modified linked list
current = modified_head
while current:
    print(current.data, end=" ")
    current = current.next
# Output: 1 3
```
