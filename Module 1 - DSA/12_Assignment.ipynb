## Question 1

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def deleteMiddleNode(head):
    if head is None or head.next is None:
        return None

    slow = head
    fast = head
    prev = None

    while fast and fast.next:
        prev = slow
        slow = slow.next
        fast = fast.next.next

    prev.next = slow.next

    return head

# Create the linked list 1->2->3->4->5
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)
head.next.next.next.next = ListNode(5)

# Call the deleteMiddleNode function
new_head = deleteMiddleNode(head)

# Print the modified linked list
current = new_head
while current:
    print(current.val, end=" ")
    current = current.next
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def findNthFromEnd(head, N):
    if head is None:
        return None

    slow = head
    fast = head

    # Move the fast pointer N-1 positions ahead
    for _ in range(N-1):
        if fast is None:
            return None
        fast = fast.next

    # If the fast pointer is None, N is out of range
    if fast is None:
        return None

    # Move both pointers until the fast pointer reaches the end
    while fast.next:
        slow = slow.next
        fast = fast.next

    # Slow pointer is at the Nth node from the end
    return slow.val
# Create the linked list based on the input values
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)
head.next.next.next.next = ListNode(5)
head.next.next.next.next.next = ListNode(6)
head.next.next.next.next.next.next = ListNode(7)
head.next.next.next.next.next.next.next = ListNode(8)
head.next.next.next.next.next.next.next.next = ListNode(9)

# Specify the value of N
N = 2

# Find the Nth node from the end
result = findNthFromEnd(head, N)

# Print the value of the Nth node from the end
print(result)  # Output: 8
```

## Question 5

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def detectAndRemoveLoop(head):
    if head is None or head.next is None:
        return

    slow = head
    fast = head

    # Detect the loop using the slow and fast pointers
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next

        if slow == fast:
            break

    # If there is no loop, return
    if slow != fast:
        return

    # Find the start of the loop
    slow = head
    while slow.next != fast.next:
        slow = slow.next
        fast = fast.next

    # Remove the loop by setting the next pointer of the last node to None
    fast.next = None

    return
# Create the linked list based on the input values
head = ListNode(1)
head.next = ListNode(3)
head.next.next = ListNode(4)

# Connect the last node to the specified position (create a loop)
X = 2  # Position at which tail is connected
tail = head
for _ in range(X-1):
    tail = tail.next

tail.next = head.next

# Remove the loop from the linked list
detectAndRemoveLoop(head)

# Print the modified linked list (without the loop)
current = head
while current:
    print(current.val, end=" ")
    current = current.next
```

## Question 7

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def mergeLists(head1, head2):
    if head2 is None:
        return head1

    curr1 = head1
    curr2 = head2

    while curr1 and curr2:
        next1 = curr1.next
        next2 = curr2.next

        curr1.next = curr2
        curr2.next = next1

        curr1 = next1
        curr2 = next2

    return head1
# Create the first linked list based on the input values
head1 = ListNode(5)
head1.next = ListNode(7)
head1.next.next = ListNode(17)
head1.next.next.next = ListNode(13)
head1.next.next.next.next = ListNode(11)

# Create the second linked list based on the input values
head2 = ListNode(12)
head2.next = ListNode(10)
head2.next.next = ListNode(2)
head2.next.next.next = ListNode(4)
head2.next.next.next.next = ListNode(6)

# Merge the second linked list into the first linked list
merged_head = mergeLists(head1, head2)

# Print the modified first linked list
current = merged_head
while current:
    print(current.val, end=" ")
    current = current.next
# Output: 5 12 7 10 17 2 13 4 11 6

# Print the modified second linked list (which should be empty)
current = head2
while current:
    print(current.val, end=" ")
    current = current.next
# Output: (no output)
```

## Question 8

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def isCircular(head):
    if head is None:
        return False

    slow = head
    fast = head.next

    while fast and fast.next:
        if slow == fast:
            return True
        slow = slow.next
        fast = fast.next.next

    return False
# Create the circular linked list based on the example
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)
head.next.next.next.next = head

# Check if the linked list is circular
result = isCircular(head)

# Print the result
print(result) 
```
