## Question 1

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None


def inorder_traversal(root, values):
    if root is None:
        return

    inorder_traversal(root.left, values)
    values.append(root.value)
    inorder_traversal(root.right, values)


def assign_sorted_values(root, values):
    if root is None:
        return

    assign_sorted_values(root.left, values)
    root.value = values.pop(0)
    assign_sorted_values(root.right, values)


def convert_to_bst(root):
    if root is None:
        return None

    values = []
    inorder_traversal(root, values)
    values.sort()
    assign_sorted_values(root, values)

    return root


def inorder_print(root):
    if root is None:
        return

    inorder_print(root.left)
    print(root.value, end=" ")
    inorder_print(root.right)


# Example usage:
# Constructing the binary tree from the input
root = TreeNode(10)
root.left = TreeNode(2)
root.right = TreeNode(7)
root.left.left = TreeNode(8)
root.left.right = TreeNode(4)

# Convert the binary tree to a binary search tree
bst_root = convert_to_bst(root)

# Print the BST in inorder traversal
print("Output:")
inorder_print(bst_root)
```

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None


def find_distance(root, node1, node2):
    if root is None:
        return 0

    if node1 < root.value and node2 < root.value:
        return find_distance(root.left, node1, node2)
    elif node1 > root.value and node2 > root.value:
        return find_distance(root.right, node1, node2)
    else:
        return distance_from_node(root, node1) + distance_from_node(root, node2)


def distance_from_node(root, node):
    if root is None:
        return 0

    if root.value == node:
        return 0
    elif node < root.value:
        return 1 + distance_from_node(root.left, node)
    else:
        return 1 + distance_from_node(root.right, node)


# Example usage:
# Constructing the binary search tree from the input
root = Node(8)
root.left = Node(3)
root.left.left = Node(1)
root.left.right = Node(6)
root.left.right.left = Node(4)
root.left.right.right = Node(7)
root.right = Node(10)
root.right.right = Node(14)
root.right.right.left = Node(13)

# Finding the distance between two nodes in the BST
node1 = 6
node2 = 14
distance = find_distance(root, node1, node2)
print("Output 1:")
print("The distance between the two keys:", distance)

node1 = 3
node2 = 4
distance = find_distance(root, node1, node2)
print("Output 2:")
print("The distance between the two keys:", distance)
```

## Question 3

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None


class DoublyLinkedListNode:
    def __init__(self, value):
        self.value = value
        self.prev = None
        self.next = None


def binary_tree_to_dll(root):
    if root is None:
        return None

    head = binary_tree_to_dll(root.left)

    new_node = DoublyLinkedListNode(root.value)
    if binary_tree_to_dll.prev is not None:
        binary_tree_to_dll.prev.next = new_node
        new_node.prev = binary_tree_to_dll.prev
    else:
        head = new_node

    binary_tree_to_dll.prev = new_node

    binary_tree_to_dll(root.right)

    return head


def print_dll(head):
    curr = head
    while curr:
        print(curr.value, end=" ")
        curr = curr.next


# Example usage:
# Constructing the binary tree from the input
root = Node(10)
root.left = Node(5)
root.right = Node(20)
root.right.left = Node(30)
root.right.right = Node(35)

# Convert the binary tree to a doubly linked list
binary_tree_to_dll.prev = None
head = binary_tree_to_dll(root)

# Print the doubly linked list
print("Output:")
print_dll(head)
```

## Question 4

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None
        self.next = None


def connect_nodes_at_same_level(root):
    if root is None:
        return

    queue = [(root, 0)]
    current_level = 0
    prev_node = None

    while queue:
        node, level = queue.pop(0)

        if level == current_level:
            if prev_node is not None:
                prev_node.next = node
            prev_node = node
        else:
            prev_node.next = None
            prev_node = node
            current_level = level

        if node.left:
            queue.append((node.left, level + 1))
        if node.right:
            queue.append((node.right, level + 1))


def print_next_pointers(root):
    node = root
    while node:
        curr = node
        while curr:
            if curr.next:
                print(curr.value, "→", curr.next.value)
            else:
                print(curr.value, "→ -1")
            curr = curr.next
        node = node.left


# Example usage:
# Constructing the binary tree from the input
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
root.right.left = Node(6)
root.right.right = Node(7)

# Connect nodes at the same level
connect_nodes_at_same_level(root)

# Print the next pointers at each level
print("Output:")
print_next_pointers(root)
```
